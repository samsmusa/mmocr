```python

from typing import List
import cv2

def make_instance(path, label):
    import os
    path = os.path.basename(path)
    print(path)
    return {"instances": [{"text": label}], "img_path": path}

def make_label_json(lines: List[str], name="textrecog", type="TextRecogDataset"):
    res = {
    "metainfo": {
        "dataset_type": type,
        "task_name": name
    },
    "data_list": []}
    for item in lines:
        line = item.strip().split(' ')
        instance = make_instance(*line)
        try:
            img = cv2.cvtColor(cv2.imread("./tests/data/dlrms_dhaka/imgs/"+instance["img_path"]), cv2.COLOR_BGR2GRAY)
            res["data_list"].append(instance)
        except:
            pass
        
    return res
```

```python
DIR_NAME = 'tests/data/dlrms_dhaka'

with open(DIR_NAME + "/word.txt", 'r') as file:
    content = file.readlines()
labels = make_label_json(content)
 
import json

with open(DIR_NAME + '/labels.json', 'w') as f:
    json.dump(labels, f)
```
